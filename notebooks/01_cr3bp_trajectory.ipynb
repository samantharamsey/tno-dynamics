# Earth–Moon CR3BP trajectory

Propagate a nondimensional trajectory in the barycentric rotating frame and examine conservation of the Jacobi constant.

```python
import matplotlib.pyplot as plt
import numpy as np

from tno_dynamics.dynamics import jacobi_constant
from tno_dynamics.propagation import propagate_cr3bp
```

## Propagation

Use the Earth–Moon mass parameter and a nonsingular initial state, sampled at 1001 evenly spaced nondimensional times.

```python
mu = 0.0121505856
state0 = np.array([0.8, 0.0, 0.0, 0.0, 0.25, 0.0])
t_eval = np.linspace(0.0, 5.0, 1001)

solution = propagate_cr3bp(state0, (0.0, 5.0), mu, t_eval=t_eval)

print(f"Integration succeeded: {solution.success}")
print(f"Solver message: {solution.message}")
print(f"State-history shape: {solution.y.shape}")
```

## Rotating-frame trajectory

Plot the trajectory with both fixed primaries and the initial state in the rotating $x$–$y$ plane.

```python
fig, ax = plt.subplots(figsize=(7, 6))
ax.plot(solution.y[0], solution.y[1], label="Trajectory")
ax.scatter(-mu, 0.0, s=90, label="Larger primary", zorder=3)
ax.scatter(1.0 - mu, 0.0, s=50, label="Smaller primary", zorder=3)
ax.scatter(state0[0], state0[1], marker="x", s=70, label="Initial state", zorder=4)
ax.set_xlabel("Rotating $x$ coordinate")
ax.set_ylabel("Rotating $y$ coordinate")
ax.set_title("Earth–Moon CR3BP trajectory in the rotating frame")
ax.set_aspect("equal", adjustable="box")
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()
```

## Jacobi-constant error

Evaluate the existing Jacobi-constant implementation along the trajectory and measure its drift from the initial value.

```python
jacobi_values = np.array([jacobi_constant(state, mu) for state in solution.y.T])
jacobi_error = jacobi_values - jacobi_values[0]

print(f"Maximum absolute Jacobi error: {np.max(np.abs(jacobi_error)):.6e}")

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(solution.t, jacobi_error)
ax.set_xlabel("Nondimensional time")
ax.set_ylabel(r"$C(t) - C(0)$")
ax.set_title("Jacobi-constant error")
ax.grid(True, alpha=0.3)
plt.show()
```
